#### 1.1.3 任务三：每1000行为一个数据模块，逐块读取（这里改为300更清晰）

```python
#写入代码
chunker = pd.read_csv('train.csv',chunksize = 300)
chunker
```

【思考】什么是逐块读取？为什么要逐块读取呢？

【提示】大家可以chunker(数据块)是什么类型？用`for`循环打印出来出处具体的样子是什么？

逐块读取：通俗的讲，把一个很大的数据文件拆分成一个个小的数据块再读入。
逐块读取的原因：减少内存资源的占用，方便对数据进行区域块处理（清洗，去重，填充）   
用for循环可以看到拆分后read5所获得的内容是多个数据块的对象（地址），而不是具体数据。

```python
for check in chunker:
  print(type(check))
  print(len(check))
  print(check.head())
```

#### 1.1.4 任务四：将表头改成中文，索引改为乘客ID [对于某些英文资料，我们可以通过翻译来更直观的熟悉我们的数据]

```python
#写入代码
df = pd.read_csv('train.csv', names=['乘客ID','是否幸存','仓位等级','姓名','性别','年龄','兄弟姐妹个数','父母子女个数','船票信息','票价','客舱','登船港口'],index_col='乘客ID',header=0)#数据表的第一行为表头
df.head()
```

【思考】所谓将表头改为中文其中一个思路是：将英文列名表头替换成中文。还有其他的方法吗？
